### Test override default na handling
### (Only Common senarios are handled, didn't use combination of upper and lower cases)

```python
na_values = [
    'N/A',
    'n/a',
    'N/a',
    'NA',
    'na',
    'NAN',
    'Nan',
    'nan',
    'NaN',
    'null',
    'Null',
    'Nil',
    'nil',
    '',
]
```

```python
df = pd.read_csv('datafortest/mixed_empties.csv', keep_default_na=False, na_values=na_values)
df.replace(r'^\s*$', np.nan, regex=True, inplace=True) # clean up white spaces
df.dropna(inplace=True)
df.head(15)
```

```python
df['orderid_is_null'] = df.isna()['Order ID']
sel = df[['Order ID', 'orderid_is_null']]
sel.head(15)
```

```python
import pandas as pd
import numpy as np

def _gen_valid_nric(x):
    nric_regex = r'^[STFG]\d{7}[A-Z]$'
    gen = rstr.xeger(nric_regex)
    return gen

df = pd.read_csv('datafortest/empties.csv')
df['NRIC'] = np.nan
df['NRIC'] = df['NRIC'].apply(lambda x: _gen_valid_nric(x))
df.head()
```
